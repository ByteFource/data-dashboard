```python
file_path = 'C:/Users/hothi/CTP MTA project/A train Analysis/MTA_Subway_Terminal_On-Time_Performance__2015-2019_20231020.csv'
df = pd.read_csv(file_path)
df
```

```python
df['date'] = pd.to_datetime(df['month'], format='%Y-%m')
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df = df[df['line'] == 'A']
df
```

```python

OnTime = df.groupby(['day_type'])['terminal_on_time_performance'].mean()
OnTime = OnTime[~OnTime.index.isin(['Systemwide'])]
OnTime = OnTime.sort_values(ascending=False)
OnTime
```

```python
OnTime = df.groupby(['year'])['terminal_on_time_performance'].mean()
OnTime = OnTime[~OnTime.index.isin(['Systemwide'])]
OnTime = OnTime.sort_values(ascending=False)
OnTime
```

```python

plt.figure(figsize=(12, 8))

plt.ylim(0, 1)
plt.gca().set_yticklabels(['{:.0%}'.format(x) for x in plt.gca().get_yticks()])
plt.xlabel('Year')
plt.ylabel('On-time Performance')
plt.title('A-Train On-time Performance from 2015-2019')
plt.xticks(df['year'].unique())
df_values = {
    2015: 0.663625,
    2016: 0.624162,
    2017: 0.549521,
    2018: 0.536486,
    2019: 0.724731
}

plt.plot(df_values.keys(), df_values.values(), color='blue', marker='o', linestyle='dashed', linewidth=2, markersize=12)

plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.gca().set_yticklabels(['{:.0f}%'.format(x*100) for x in plt.gca().get_yticks()])
plt.xlabel('Year')
plt.ylabel('On-time Performance')
plt.title('A-Train On-time Performance from 2015-2019')
plt.xticks(df['year'].unique())
df.groupby(['year'])['terminal_on_time_performance'].mean().plot(kind='line', color='red', marker='o', linestyle='dashed', linewidth=2, markersize=12)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.title('A train Heatmap for On-Time Performance from 2015-2019')
sns.heatmap(df.corr(), annot=True, cmap='PiYG', vmin=-1, vmax=1, linewidths=0.2, annot_kws={'size': 12}, fmt='.2f')
plt.show()
```
